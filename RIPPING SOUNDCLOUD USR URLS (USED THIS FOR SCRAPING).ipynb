I was working for a Music project that needed to find new listeners for its music. I made this in order to find users who like similar music to what we were producing. This Notebook will take a songs Soundcloud page and scrape all of the usernames of users who have commented on the track. The idea was that users who comment on a song are interacting and would good good followers to have. 

```python
from selenium import webdriver
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Firefox()
```

The track I used as a test case was Flume - Never Be Like You

```python

driver.get("https://soundcloud.com/flume/never-be-like-you-feat-kai")
```

```python
html = driver.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html)
```

```python
print soup.prettify()
```

Search for all links on the page to find all users names of commentors.

```python
all_links= soup.find_all('a')
```

```python
print all_links
```

finds all 'a' class .... hrefs  and puts in a list 

```python

usr_url_list= []
for a in soup.findAll('a', {'class':'commentItem__usernameLink sc-link-light'}):
    usr_url_list.append(a['href'])
    #print "Found the URL:", a['href']
```

List of Usernames that were then used in the "FASTER FOLLOWER CHECK VIA URLIB" NoteBook.
